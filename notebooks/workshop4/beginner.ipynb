## Step 1 · Ask your assistant
Open your assistant in another tab and paste this prompt:

> Write a Python function `count_up(path)` that reads a DESeq2 results CSV with pandas (gene IDs are in
> the first column), and returns how many genes have `log2FoldChange > 1` **and** `padj < 0.05`.

It'll hand you back a function. Paste it into the next cell, replacing the version we pre-filled.

```python
# --- paste your assistant's function over this one, then run the cell ---
# (this is a correct reference so the notebook works even before you paste)
def count_up(path):
    df = pd.read_csv(path, index_col=0)
    hits = df[(df['log2FoldChange'] > 1) & (df['padj'] < 0.05)]
    return len(hits)
```

## Step 2 · Verify it against what you know
Now the whole point. Run the function on bone marrow and compare its answer to **1343** — the number *you*
produced last workshop. If they don't match, the code is wrong, no matter how good it looks.

```python
n = count_up(os.path.join(DATA, 'de', 'BM_DE.csv'))
print('the assistant\'s function says:', n, 'genes up in bone marrow')
known = ...   # the number you derived in Workshop 3
print('MATCH ✓' if n == known else f'MISMATCH ✗  (off by {n - known:+d})')
```
